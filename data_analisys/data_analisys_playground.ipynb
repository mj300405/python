## Importowanie danych z pliku **CSV**

```python
import pandas as pd

data=pd.read_csv("dane.csv")
data['droga_hamowania']=data['droga_hamowania'].abs()
x=data['predkosc']
y=data['droga_hamowania']
```

## Wykres danych, regresja liniowa

```python
import matplotlib.pyplot as plt
import numpy as np
slope,inter=np.polyfit(x,y, 1)
x1=np.linspace(0,200, 5)
y1=x1*slope+inter
plt.scatter(x,y)
plt.plot(x1,y1, color='red')
plt.grid()
plt.xlabel('prędkość')
plt.ylabel('droga hamowania')
plt.title('Zależność drogi od czasu')
plt.show()
```

## Wykres danych, regresja wykładnicza

```python
import matplotlib.pyplot as plt
import numpy as np
exponential_fit=np.polyfit(x,np.log(y), 1, w=np.sqrt(y))
x3=np.linspace(x.min(), x.max(), 100)
y3=np.exp(exponential_fit[1])*np.exp(exponential_fit[0]*x3)
plt.scatter(x,y)
plt.plot(x3,y3,'r')
plt.show()
```

## Wykres danych, regresja kwadratowa

```python
import numpy as np
import matplotlib.pyplot as plt
quadratic_fit=np.polyfit(x,y,2)
x2=np.linspace(x.min(),x.max(),100)
y2=np.polyval(quadratic_fit, x2)
plt.scatter(x,y)
plt.plot(x2,y2, 'r')
plt.show()
```
